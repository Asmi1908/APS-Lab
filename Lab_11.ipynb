# Travelling Salesman Problem using Dynamic Programming.
Problem Description:
Given a set of cities and distances, find the shortest possible route that visits each city once and returns to the starting city.

#Theory
TSP uses dynamic programming with bit masking. Store results of visited subsets to reduce repeated work.

## Algorithm
1. Start from source city
2. Mark visited cities using bit mask
3. Try all unvisited cities
4. Choose minimum cost path
5. Return to source city

```python
def tsp(graph):
    n = len(graph)
    dp = [[-1]*(1<<n) for _ in range(n)]
    path_tracker = [[-1]*(1<<n) for _ in range(n)] # To store the next city for the optimal path

    def visit(city, mask):
        if mask == (1<<n) - 1:
            # All cities visited, return to the starting city (city 0)
            path_tracker[city][mask] = 0 # Mark that the next step is to return to city 0
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')
        best_next_city = -1

        for next_city_cand in range(n):
            # Check if next_city_cand has not been visited yet and is not the current city
            if not (mask & (1<<next_city_cand)):
                new_cost = graph[city][next_city_cand] + visit(next_city_cand, mask | (1<<next_city_cand))
                if new_cost < ans:
                    ans = new_cost
                    best_next_city = next_city_cand # Store the next city that leads to this minimum cost

        dp[city][mask] = ans
        path_tracker[city][mask] = best_next_city # Store the decision
        return ans

    min_cost = visit(0, 1) # Start from city 0, mask 1 (0001 - city 0 visited)

    # Reconstruct the path
    current_city = 0
    current_mask = 1
    optimal_path_cities = [0] # Start at city 0

    while True:
        next_city = path_tracker[current_city][current_mask]
        # Break condition: when we are at the last visited city and the next step is to return to 0
        if next_city == 0 and current_mask == (1<<n) - 1:
            break
        optimal_path_cities.append(next_city)
        current_mask |= (1 << next_city)
        current_city = next_city

    optimal_path_cities.append(0) # Add the final return to city 0

    return min_cost, optimal_path_cities
```

```python
graph = [
[0, 10, 15, 20],
[10, 0, 35, 25],
[15, 35, 0, 30],
[20, 25, 30, 0]
]

min_cost, path = tsp(graph)
print("Minimum travelling cost:", min_cost)
print("Optimal path:", path)
```

```python
def get_user_graph():
    while True:
        try:
            n = int(input("Enter the number of cities: "))
            if n <= 1:
                print("Number of cities must be at least 2.")
                continue
            break
        except ValueError:
            print("Invalid input. Please enter an integer.")

    graph = []
    print("Enter the distance matrix row by row. Enter 0 for distance from a city to itself.")
    print("Example for 3 cities: 0,10,15")

    for i in range(n):
        while True:
            try:
                row_str = input(f"Enter distances for city {i} (comma-separated, {n} values): ")
                row_values = list(map(int, row_str.split(',')))
                if len(row_values) != n:
                    print(f"Please enter exactly {n} values.")
                    continue
                if row_values[i] != 0:
                    print(f"Distance from city {i} to itself must be 0.")
                    continue
                # Optional: Add checks for symmetry if it's an undirected graph
                graph.append(row_values)
                break
            except ValueError:
                print("Invalid input. Please enter comma-separated integers.")
    return graph
```

Now, let's use the `get_user_graph()` function to create a graph based on your input and then solve the TSP for it.

```python
# Get graph input from the user
user_graph = get_user_graph()

# Solve TSP for the user-provided graph
min_cost_user, path_user = tsp(user_graph)

print("\n--- User Defined Graph TSP Result ---")
print("Minimum travelling cost:", min_cost_user)
print("Optimal path:", path_user)
```
