# Warshall algorithm to find reachability between all pairs of vertices in a graph.
Problem Description:
Understand transitive closure of a graph. Use Python to compute reachability using matrix method.

#Theory
Warshall algorithm finds whether a path exists between every pair of vertices. It updates the adjacency matrix step by step. If there is a path from i to j through k, then the entry becomes 1.

## Algorithm
1. Represent graph using adjacency matrix.
2. For each intermediate vertex k
3. Check all pairs i and j
4. Update value if path exists through k
4. Repeat for all vertices


```python
def warshall(graph):

    n = len(graph)

    for k in range(n):
        for i in range(n):
            for j in range(n):
                graph[i][j] = graph[i][j] or (graph[i][k] and graph[k][j])

    return graph
```

```python
n = int(input("Enter the number of vertices: "))
graph = []
print("Enter the adjacency matrix row by row (0 or 1, space-separated):")
for i in range(n):
    row_input = input(f"Enter row {i + 1}: ").split()
    row = [int(x) for x in row_input]
    graph.append(row)

result = warshall(graph)

print("Transitive Closure Matrix (Reachability):")

for i in range(n):
    for j in range(n):
        if result[i][j] == 1:
            print(f"There is a path from vertex {i} to vertex {j}")
        else:
            print(f"There is NO path from vertex {i} to vertex {j}")
```

## Find shortest paths between all pairs of vertices using Floyd Warshall algorithm.

#Understand all pairs shortest path problem. Implement dynamic programming approach in Python.

* Given a weighted graph with positive or negative edge weights, find
shortest distances between every pair of vertices.

# Theory
Floyd Warshall algorithm computes shortest paths between all pairs of nodes. It updates the distance matrix by checking if a shorter path exists through an intermediate vertex. Works with negative weights but fails with negative cycles.

Algorithm
1. Represent graph using distance matrix.
2. Initialize matrix with given weights.
3. Set distance of node to itself as 0.
4. For each intermediate vertex k
5. Update distance[i][j] = min(distance[i][j], distance[i][k] + distance[k][j])
6. Repeat for all vertices

```python
def floyd_warshall(graph):

    n = len(graph)

    dist = []
    for i in range(n):
        dist.append(graph[i][:])

    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][k] + dist[k][j] < dist[i][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]

    return dist

```

```python
INF = float('inf')

n = int(input("Enter the number of vertices: "))
print("Enter the distance matrix row by row (use 'inf' for no direct edge, space-separated):")

graph = []
for i in range(n):
    row_input = input(f"Enter row {i + 1}: ").split()
    row = []
    for x in row_input:
        if x.lower() == 'inf':
            row.append(INF)
        else:
            row.append(int(x))
    graph.append(row)

result = floyd_warshall(graph)

print("Shortest distance matrix:")

for row in result:
    print(row)
```

```python
negative_cycle_detected = False
for i in range(n):
    if result[i][i] < 0:
        negative_cycle_detected = True
        break

if negative_cycle_detected:
    print("Negative cycle detected in the graph.")
else:
    print("No negative cycle detected in the graph.")
```

```python
start_vertex = int(input("Enter the starting vertex (0-indexed): "))
end_vertex = int(input("Enter the ending vertex (0-indexed): "))

if 0 <= start_vertex < n and 0 <= end_vertex < n:
    shortest_distance = result[start_vertex][end_vertex]
    if shortest_distance == INF:
        print(f"There is no path from vertex {start_vertex} to vertex {end_vertex}.")
    else:
        print(f"The shortest distance from vertex {start_vertex} to vertex {end_vertex} is: {shortest_distance}")
else:
    print("Invalid vertex indices. Please enter values within the graph's range.")
```
