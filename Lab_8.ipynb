# Minimum Cost Spanning Tree
Problem Description:
A spanning tree of a graph can be defined as a graph with minimal set of edges that connect all vertices. A minimum spanning tree of a graph is a spanning tree of the graph with least weight (where the weight is computed by adding the weights of all the edges in the spanning tree). In general, a graph can have multiple minimum spanning trees. The problem is to find a minimum spanning tree of a graph.

This implementation uses a dictionary to represent the weighted graph, where the vertices are keys, and the values are dictionaries containing their adjacent vertices and their weights.

Please note that this implementation assumes that the input graph is connected and weighted.

First, let's create a simple graph to demonstrate Prim's algorithm:

```python
graph = {
    "A": {"B": 10, "C": 15, "D": 20},
    "B": {"A": 10, "C": 12},
    "C": {"A": 15, "B": 12, "D": 18},
    "D": {"A": 20, "C": 18},
}
```

Now, let's implement Prim's algorithm:

```python
def prim_algorithm(graph):
    min_tree_edges = []
    min_tree_vertices = set()
    min_weight = 0

    start_vertex = list(graph.keys())[0]
    min_tree_vertices.add(start_vertex)

    while len(min_tree_vertices) < len(graph):
        min_edge_weight = float('inf')
        min_edge_u = None
        min_edge_v = None

        for u in min_tree_vertices:
            for v, weight in graph[u].items():
                if v not in min_tree_vertices:
                    if weight < min_edge_weight:
                        min_edge_weight = weight
                        min_edge_u = u
                        min_edge_v = v

        if min_edge_v is not None:
            min_tree_edges.append((min_edge_u, min_edge_v, min_edge_weight))
            min_tree_vertices.add(min_edge_v)
            min_weight += min_edge_weight
        else:
            break

    return min_tree_edges, min_weight
```

Now, let's run the Prim's algorithm on the example graph:
Keep in mind that this example uses a simple representation of a graph, but for more complex cases, you might want to use a library such as NetworkX to represent and manipulate graphs.

```python
mst_edges, min_weight = prim_algorithm(graph)
print("Minimum Spanning Tree Edges: ", mst_edges)
print("Minimum Weight: ", min_weight)
```

Use same apporach for Kruskal's algorithm

```python
class DSU:
    def __init__(self, vertices):
        self.parent = {v: v for v in vertices}
        self.rank = {v: 0 for v in vertices}

    def find(self, i):
        if self.parent[i] == i:
            return i
        self.parent[i] = self.find(self.parent[i])
        return self.parent[i]

    def union(self, i, j):
        root_i = self.find(i)
        root_j = self.find(j)
        if root_i != root_j:
            if self.rank[root_i] < self.rank[root_j]:
                self.parent[root_i] = root_j
            elif self.rank[root_i] > self.rank[root_j]:
                self.parent[root_j] = root_i
            else:
                self.parent[root_j] = root_i
                self.rank[root_i] += 1
            return True
        return False

def kruskal_algorithm(graph):
    edges = []
    for u in graph:
        for v, weight in graph[u].items():
            # Add each edge once to avoid duplicates (e.g., (A,B) and (B,A))
            if (v, u, weight) not in edges:
                edges.append((u, v, weight))

    # Sort edges by weight
    edges.sort(key=lambda x: x[2])

    vertices = list(graph.keys())
    dsu = DSU(vertices)
    mst_edges = []
    min_weight = 0

    for u, v, weight in edges:
        if dsu.union(u, v):
            mst_edges.append((u, v, weight))
            min_weight += weight
            if len(mst_edges) == len(vertices) - 1:
                break

    return mst_edges, min_weight

# Run Kruskal's algorithm on the example graph
mst_kruskal, min_weight_kruskal = kruskal_algorithm(graph)
print("Minimum Spanning Tree (Kruskal's): ", mst_kruskal)
print("Minimum Weight (Kruskal's): ", min_weight_kruskal)
```
